```python
# Create a a list of random numbers
random_num = np.random.random_integers(1, 26000, 1000)
```

```python
random_num = list(set(random_num))
print(len(random_num))
```

```python
# Lists for holding IDs
cities = []

# Create a Loop 
for i in range(len(random_num)):
    base_url = requests.get(f'https://www.ncdc.noaa.gov/cdo-web/api/v2/locations?offset={random_num[i]}&limit=1',headers={'token':api_key})
    d = json.loads(base_url.text)
    cities_id = d['results'][0]['id']
    
    # for each index append city ID
    cities.append(cities_id)
    print(i)
```

```python
cities
```

```python
# Set up a join for all cities
city_list = "&".join(cities)
city_list
```

```python
# Looping through cities
for city in cities:
    for year in range(1980, 2010, 10):
        start_year = year
        end_year = year+9
        locations = requests.get(f'https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GSOY&locationid={city}&limit=1000&startdate={start_year}-01-01&enddate={end_year}-12-31', headers={'token':api_key})
        loc = json.loads(locations.text)
        if bool(loc):
            json_file = f"Resources/{city}_{start_year}.json"
            json.dump(loc, open(json_file, "w+"))
            iterration = ceil(int(loc['metadata']['resultset']['count'])/1000)-1
            for i in range(1, iterration):
                offset = (1000*i)+1
                locations = requests.get(f'https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GSOY&locationid={city}&limit=1000&startdate={start_year}-01-01&enddate={end_year}-12-31&offset={offset}', headers={'token':api_key})
                loc = json.loads(locations.text)
                json_file = f"Resources/{city}_{start_year}_{i}.json"
                json.dump(loc, open(json_file, "w+"))

          
```
